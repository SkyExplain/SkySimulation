```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

nside = np.array([64, 128, 254, 512, 1024])
time = np.array([20, 130, 7000, 25000, 100000])

plt.plot(nside, time, 'o--')
plt.xlabel('Nside')
plt.ylabel('Time (s)')
plt.grid(color='lightgray', linestyle='--')
plt.show()
```
